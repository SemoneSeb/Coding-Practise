```
n = int(input("Enter the no. of elements in the array : "))
if n<3:
  print("Invalid no. of values")
arr = []
p_arr = []
for i in range (n):
  a = int(input())
  arr.append(a)
print(arr)
for x in range (0,n-2):
  for y in range (x+1, n-1):
    for z in range (y+1, n):
      a = arr[x]*arr[y]*arr[z]
      p_arr.append(a)
result = max(p_arr) 
print(result)
```
